#### 1-recursion algorithm

```python
#recursion

def recur(n: int) -> int:
    #终止条件-stop condition
    if n == 1:
        return 1
    #递：递归调用-recursion part
    res = recur(n-1)
    #归：返回结果-return part
    return n + res
```

```python
recur(5)
```

#### 2-calculate determinat by recursion algorithm

```python
# to calculate determinat by recursion algorithm
 
Mtx1 = [[1,7,2,4],[1,5,2,4],[3,0,1,0],[2,1,5,-3]]
Mtx1
```

```python
def minor(mtx,row):
    
    new_mtx = []

    for i in range(len(mtx)):
        #print("i = ",i)
            
        if i != row:
            #print("row = ",i)
            #print(mtx[i][1:])
            new_mtx.append(mtx[i][1:])
    #print("*"*48)
    return new_mtx


def dtm(mtx):

    #终止条件-stop condition
    if len(mtx)==2 and len(mtx[0])==2:
        return mtx[0][0]*mtx[1][1] - mtx[0][1]*mtx[1][0]
    
    det = 0
    for row in range(len(mtx)):
        a_i = mtx[row][0]                # fix column at the first column (col = 0)
        minor_i = minor(mtx,row)          # minor matrix excluding row_i, col_0
        #print(f"a_i = a_{row} =", a_i)
        #print(f"minor_i = minor matrix excluding row_{row}, col_0")
    
        #递：递归调用-recursion part
        cofactor = a_i * dtm(minor_i)
    
        det += (-1)**(row)*cofactor

    #归：返回结果-return part
    return det

```
